```python
sequences = {}
current_id = ""
with open("rosalind_cons.txt") as f:
    for line in f:
        line = line.strip()
        if line.startswith(">"):
            current_id = line[1:]  
            sequences[current_id] = ""
        else:
            sequences[current_id] += line
n = len(list(sequences.values())[0])
profile = {'A': [0]*n, 'C': [0]*n, 'G': [0]*n, 'T': [0]*n}
for seq in sequences.values():
    for i,base in enumerate(seq):
        profile[base][i]+=1

consensus = ""
for i in range(n):
    consensus+=max('ACGT', key=lambda b: profile[b][i])
print(consensus)
for base in 'ACGT':
    print(f"{base}: {' '.join(map(str, profile[base]))}")
```

```python
seq = {}
current_id = ""
with open("rosalind_lcsm.txt") as f:
    for line in f:
        line = line.strip()
        if line.startswith(">"):
            current_id = line[1:]
            seq[current_id] = ""
        else:
            seq[current_id] += line
shortest = min(seq.values(), key=len)
all_sequences = list(seq.values())

longest = ""
for i in range(len(shortest)):
    for j in range(i+1, len(shortest)+1):
        substring = shortest[i:j]
        if all(substring in s for s in all_sequences):
            if len(substring) > len(longest):
                longest = substring

print(longest)
```
